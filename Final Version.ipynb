```python
effects = []
for element in names:
    parts = element.split(',')
    effects.append(parts)

more = []
for n in effects:
    for i in n:
            if i not in effects:
                more.append(i)
```

```python
for i in more:
    title=str(i)
    title=[]
    for x in df.effects:
        if i in x:
            title.append(1)
        else:
            title.append(0)
    df[i]=title
```

```python
tastes = list(df.taste.values)
taste = []
for element in tastes:
    parts = element.split(',')
    taste.append(parts)

mores = []
for n in taste:
    for i in n:
            if i not in taste:
                mores.append(i)
for i in mores:
    title=str(i)
    title=[]
    for x in df.taste:
        if i in x:
            title.append(1)
        else:
            title.append(0)
    df[i]=title
```
